```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie=pd.read_csv('movie.csv')
rating=pd.read_csv('rating.csv')
rating.drop('timestamp',axis=1,inplace=True)
movie.drop('genres',axis=1,inplace=True)
rating.head(2)
movie.head(2)
```

```python
df=pd.merge(movie,rating)
df=df[df['movieId']<10000]
df=df[df['userId']<10000]
df.head(1)
```

```python
df.head()
```

```python
ds=df.pivot_table(index=['userId'],columns=['title'],values='rating').dropna(thresh=500,axis=1)
ds.head(2)
```

```python
ds.fillna(0,inplace=True)
ds.head()
```

```python
movie_similarity = ds.corr(method='pearson')
movie_similarity.head()
```

```python
def get_similar_movies(movie_name,user_rating):
    similarity_score=movie_similarity[movie_name]
    similarity_score=similarity_score.sort_values(ascending=False)
    return similarity_score

print(get_similar_movies("Toy Story (1995)",1))
```
